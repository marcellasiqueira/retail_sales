###Testes Estatísticos de Diferença de Vendas por Gênero e Categoria (Teste T)

Esta célula realiza testes t independentes para comparar o total de vendas entre os gêneros feminino e masculino dentro de cada categoria de produto, a fim de verificar se as diferenças observadas são estatisticamente significativas.

- `from scipy.stats import ttest_ind`: Importa a função `ttest_ind` do módulo `scipy.stats` para realizar o teste t independente.
- O código itera sobre cada categoria de produto única.
- Para cada categoria, filtra o DataFrame para obter as vendas totais para o gênero feminino e masculino separadamente.
- `ttest_ind(sales_female, sales_male)`: Realiza o teste t comparando as amostras de vendas femininas e masculinas para a categoria atual. Retorna a estatística t e o valor p.
- O código imprime a estatística t, o valor p e uma conclusão sobre a significância estatística (se p < 0.05).

```python
from scipy.stats import ttest_ind

product_categories = df_sales['Product Category'].unique()

for category in product_categories:
    sales_female = df_sales[(df_sales['Product Category'] == category) & (df_sales['Gender'] == 'Female')]['Total Amount']
    sales_male = df_sales[(df_sales['Product Category'] == category) & (df_sales['Gender'] == 'Male')]['Total Amount']

    if len(sales_female) > 1 and len(sales_male) > 1: # Ensure there are enough samples for the test
        t_stat, p_value = ttest_ind(sales_female, sales_male)
        print(f"Category: {category}")
        print(f"  T-statistic: {t_stat:.4f}")
        print(f"  P-value: {p_value:.4f}")
        if p_value < 0.05:
            print("  Result: Statistically significant difference")
        else:
            print("  Result: No statistically significant difference")
    else:
        print(f"Category: {category}")
        print("  Not enough samples to perform t-test")
```

###Análise de Associação entre Gênero e Categoria (Tabela de Contingência)

Esta célula cria uma tabela de contingência para mostrar a distribuição conjunta das variáveis categóricas 'Product Category' e 'Gender'. Esta tabela é a base para o teste Qui-quadrado de independência.

- `contingency_table = pd.crosstab(df_sales['Product Category'], df_sales['Gender'])`: Utiliza a função `pd.crosstab()` para criar a tabela de contingência, onde as linhas são as categorias de produto e as colunas são os gêneros. Os valores na tabela são as contagens de ocorrências para cada combinação.
- `display(contingency_table)`: Exibe a tabela de contingência.

###Análise de Associação entre Gênero e Categoria (Teste Qui-quadrado)

Esta célula realiza um teste Qui-quadrado de independência na tabela de contingência criada anteriormente para determinar se há uma associação estatisticamente significativa entre Gênero e Categoria de Produto.

- `from scipy.stats import chi2_contingency`: Importa a função `chi2_contingency` do módulo `scipy.stats`.
- `chi2, p, dof, expected = chi2_contingency(contingency_table)`: Realiza o teste Qui-quadrado na tabela de contingência. Retorna a estatística Qui-quadrado, o valor p, os graus de liberdade e as frequências esperadas.
- O código imprime a estatística Qui-quadrado, o valor p e os graus de liberdade.
- Com base no valor p (comparado com 0.05), o código imprime uma conclusão sobre se há ou não uma associação estatisticamente significativa entre Gênero e Categoria de Produto.

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-squared statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
# print("Expected frequencies:")
# display(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

if p < 0.05:
    print("Result: Statistically significant association between Gender and Product Category")
else:
    print("Result: No statistically significant association between Gender and Product Category")
```

###Testes Estatísticos de Diferença de Ticket Médio por Categoria (ANOVA)

Esta célula realiza um teste de Análise de Variância (ANOVA) de mão única para verificar se há uma diferença estatisticamente significativa no ticket médio entre as diferentes categorias de produto.

- `from scipy import stats`: Importa o módulo `stats` da biblioteca SciPy.
- O código cria listas separadas com os valores de 'Total Amount' para cada categoria de produto.
- `stats.f_oneway(beauty_sales, clothing_sales, electronics_sales)`: Realiza o teste ANOVA comparando as médias das vendas totais entre as três categorias. Retorna a estatística F e o valor p.
- O código imprime a estatística F do ANOVA, o valor p e uma conclusão sobre a significância estatística (se p < 0.05) da diferença nos tickets médios entre as categorias.

```python
from scipy import stats

# Create a list of sales amounts for each product category
beauty_sales = df_sales[df_sales['Product Category'] == 'Beauty']['Total Amount']
clothing_sales = df_sales[df_sales['Product Category'] == 'Clothing']['Total Amount']
electronics_sales = df_sales[df_sales['Product Category'] == 'Electronics']['Total Amount']

# Perform one-way ANOVA test
f_statistic, p_value = stats.f_oneway(beauty_sales, clothing_sales, electronics_sales)

print(f"ANOVA F-statistic: {f_statistic:.4f}")
print(f"ANOVA P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: Statistically significant difference in average ticket between categories")
else:
    print("Result: No statistically significant difference in average ticket between categories")
```

O eixo X são os dias da semana (na ordem categórica), o eixo Y o número de transações, e as linhas são coloridas ('hue') pela categoria de produto. `marker='o'` adiciona marcadores nos pontos de dados.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Convert 'Day of Week' to a categorical type with the specified order
sales_by_day_category['Day of Week'] = pd.Categorical(sales_by_day_category['Day of Week'], categories=day_order, ordered=True)

# Sort the DataFrame by the ordered 'Day of Week'
sales_by_day_category = sales_by_day_category.sort_values('Day of Week')


plt.figure(figsize=(12, 6))
sns.lineplot(x='Day of Week', y='Number of Transactions', hue='Product Category', data=sales_by_day_category, marker='o')
plt.title('Number of Transactions by Day of Week and Product Category')
plt.xlabel('Day of Week')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
